```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
%matplotlib inline
import numpy as np
import time
from time import sleep
```

```python
soups= []
main_url= 'https://www.news247.gr/volontimir-zelenski?pages='
for i in range(1,6):
  latest_page_url= main_url + str(i)
  print (latest_page_url)
  response= requests.get(latest_page_url)
  soup= bs(response.text, 'html.parser') 
  soups.append(soup) 
  time.sleep(2)
```

```python
soups[0]
```

```python
urls_list= []
for soup in soups: 
  articles_list=soup.find_all('article', {'class':'teaser__article article article_left '}) 
  for article in articles_list: 
    url= article.find('h2').find('a')['href'] 
    urls_list.append(url)
```

```python
urls_list
```

```python
len(urls_list)
```

```python
full_articles_list_html=[]
for url in urls_list:
  response= requests.get(url)
  soup= bs(response.text, 'html.parser')
  full_articles_list_html.append(soup)
```

```python
final_list= []
for html in full_articles_list_html:
    final_dict = {}
    titl = html.find('div', {'class': 'article-body '}).find('h1')
    if titl:
        final_dict['title'] = titl.text
    lead = html.find('h2', {'class': 'article-body__introtext'})
    if lead:
        final_dict['lead'] = lead.text
    section_tag = html.find('div', {'class': 'article-head'}).find('a')
    if section_tag:
        final_dict['section_tag'] = section_tag.text 
    fig= html.find('figure', {'class':'article-body__picture'}) 
    if fig: 
        img = fig.find('img') 
        if img:
            final_dict['Alternative_text'] = img['alt'] 
            final_dict['Image_url'] = img['src'] 
    
    date= html.find('time')
    if date:
        final_dict['date'] = date['datetime']
    
    body = []
    for p in html.find('div', {'class': 'article-body__storyContent '}).find_all(['p', 'h2']):
      body_text = p.text 
      body.append(body_text)
    if body_text: 
      full_body_text =' '.join(body).replace("  ", " ") 
      final_dict['body'] = full_body_text 
    
    editor = html.find('div', {'class': 'article-body__author-name'}) 
    if editor:                                                        
      final_dict['editor'] = "".join(editor.text.splitlines())        
    else:                                                            
      editor = html.find('div', {'class': 'article-body__byline'}).find('a') 
      if editor:                                                      
        final_dict['editor'] = "".join(editor.text.splitlines())      
      

    final_list.append(final_dict)
df_Zelensky = pd.DataFrame(final_list)
df_Zelensky
```

```python
df_Zelensky['title'] = df_Zelensky['title'].str.replace(r'[\r\n\t]+',"")
df_Zelensky['section_tag'] = df_Zelensky['section_tag'].str.replace(r'[\r\n\t]+',"")
df_Zelensky['editor'] = df_Zelensky['editor'].str.replace(r'[\r\n\t]+',"")
df_Zelensky
```
